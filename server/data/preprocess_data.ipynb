```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_json('leetcode-solution.jsonl', lines=True)
```

```python
df
```

```python
# Set randomly the difficulty, easy in range (1000- 3000), medium in range (3000 - 6000), hard in range (6000 - 10000)
def set_difficulty(row):
    if row['difficulty'] == 'Easy':
        return np.random.randint(1000, 3000)
    elif row['difficulty'] == 'Medium':
        return np.random.randint(3000, 6000)
    else:
        return np.random.randint(6000, 10000)
df['difficulty'] = df.apply(set_difficulty, axis=1)
```

```python
df = df[['title', 'difficulty', 'content']]
```

```python
# Extract description, examples, and constraints from content
import re

def extract_content_parts(content):
    # Extract description (everything before the first "**Example")
    description_match = re.search(r'^(.*?)(?=\*\*Example|\*\*Constraints:)', content, re.DOTALL)
    description = description_match.group(1).strip() if description_match else ""
    
    # Extract examples (from "**Example" to "**Constraints:")
    examples_section = re.search(r'(\*\*Example.*?)(?=\*\*Constraints:)', content, re.DOTALL)
    examples_text = examples_section.group(1) if examples_section else ""
    
    # Format examples to extract only Input and Output
    formatted_examples = []
    if examples_text:
        # Find all Input/Output pairs
        input_output_pairs = re.findall(r'\*\*Input:\*\*(.*?)\*\*Output:\*\*(.*?)(?=\*\*|$)', examples_text, re.DOTALL)
        
        for input_val, output_val in input_output_pairs:
            # Clean up the input and output values
            input_clean = re.sub(r'\s+', ' ', input_val.strip())
            output_clean = re.sub(r'\s+', ' ', output_val.strip())
            
            # Remove explanation parts (everything after "**Explanation:")
            # output_clean = re.sub(r'\*\*Explanation:.*', '', output_clean).strip()
            
            formatted_examples.append(f"Input: {input_clean}\n\nOutput: {output_clean}")
    
    examples = '\n\n'.join(formatted_examples)
    
    # Extract constraints (from "**Constraints:" to the end)
    constraints_match = re.search(r'\*\*Constraints:\*\*(.*)', content, re.DOTALL)
    constraints = constraints_match.group(1).strip() if constraints_match else ""
    
    # Remove backticks from constraints
    constraints = re.sub(r'`([^`]*)`', r'\1', constraints)
    
    # Remove asterisks (*) from constraints
    constraints = re.sub(r'\*', '', constraints)
    # Remove extra spaces but keep newlines
    constraints = re.sub(r'[ \t]+', ' ', constraints).strip()
    return description, examples, constraints

# Apply the extraction function to create new columns
df[['description', 'examples', 'constraints']] = df['content'].apply(
    lambda x: pd.Series(extract_content_parts(x))
)
```

```python
# Display the extracted parts for verification
print("Description:")
print(df.iloc[0]['description'])
print("\n" + "="*50 + "\n")

print("Examples:")
print(df.iloc[0]['examples'])
print("\n" + "="*50 + "\n")

print("Constraints:")
print(df.iloc[0]['constraints'])
```
